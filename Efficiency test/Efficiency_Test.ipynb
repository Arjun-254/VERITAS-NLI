```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import time


def people_also_ask(url):
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument('log-level=1')

    driver = webdriver.Chrome(options=options)
    driver.get(url)
    try:
        wait = WebDriverWait(driver, 5)
        questions = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@jsname='yEVEwb']")))
    except Exception as e:
        # print(f"People Also Asked questions dont exist for this category")
        driver.close()
        driver.quit()

        return []

    all_span_texts = []

    for div in questions:
        try:
            div.click()
            time.sleep(1)

            span_elements = div.find_elements(By.TAG_NAME, 'span')
            for span in span_elements:
                text = span.text.strip()
                if text and text not in all_span_texts:
                    all_span_texts.append(text)

        except Exception as e:
            print(f"Error clicking or extracting from div/span: {e}")
    driver.close()
    driver.quit()

    return all_span_texts[0:2]

```

```python
def get_top_links(query, num_links=1):
    linksgot = []
    try:
        # Perform Google search and get the top links
        search_results = search(query, num_results=num_links)

        # Print the top links
        for i, link in enumerate(search_results, start=1):
            linksgot.append(link)
    except Exception as e:
        print(f"An error occurred: {e}")
    return linksgot


def get_title_and_content(search_query_results):
    article_titles = []
    article_content = []

    # Set up Selenium options (headless mode if needed)
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument('log-level=1')

    driver = webdriver.Chrome(options=options)

    if search_query_results:
        for result in search_query_results:
            try:
                # Navigate to the URL
                driver.get(result)

                # Scrape <h1> tags for titles
                h1_tags = driver.find_elements(By.TAG_NAME, 'h1')
                currenth1 = " ".join([h1.text for h1 in h1_tags])
                article_titles.append(currenth1)

                # Scrape <p> tags for content
                p_tags = driver.find_elements(By.TAG_NAME, 'p')
                currentp = " ".join([p.text for p in p_tags])
                article_content.append(currentp)

            except Exception as e:
                print(f"An error occurred: {e}")

    # Quit the driver'
    driver.close()
    driver.quit()

    return article_titles, article_content

# Get the titles and contents
def make_data(search_query_results):
    titles, contents = get_title_and_content(
        get_top_links(search_query_results))

    # Create a pandas DataFrame
    data = {'Title': titles, 'Content': contents}
    df = pd.DataFrame(data)
    return df
```

```python
#To answer the question generated by SLM via google
def google_search(query):
    headers = {
        'User-agent':
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'
    }

    # Perform the Google search
    query = re.sub(r'[^a-zA-Z0-9.,!?;:()\- ]+', '', query)
    search_url = f'https://www.google.com/search?q={query}'
    html = requests.get(search_url, headers=headers)

    # Parse the HTML response
    soup = BeautifulSoup(html.text, 'html.parser')

    #extract answer from summary answer on google
    if soup.select_one('.hgKElc'):
        answer = soup.select_one('.hgKElc').text
        print('\n',answer)

    # Extract the answer (assuming it's in a specific class) -> QUICK ANSWER BOX
    elif soup.select_one('.DI6Ufb'):
        answer = soup.select_one('.DI6Ufb')
        answer = answer.find(class_='Z0LcW t2b5Cf').text
        answer=query+' '+answer

    elif not soup.select_one('.DI6Ufb'):
        result=people_also_ask(search_url)
        # print("Result of the Scraping",result)
        if len(result)==2:
            answer=f'{result[0]} {result[1]}'
        else:
            #incase people also asked also comes empty we revert to the article pipeline
            scraped_df = make_data(query)
            scraped_df.dropna(inplace=True)

            # Convert the 'Content' column to strings
            scraped_df['Content'] = scraped_df['Content'].astype(str)

            # Sort the DataFrame based on the length of the strings in the 'Content' column
            scraped_df = scraped_df.sort_values(by='Content', key=lambda x: x.str.len(), ascending=False)

            if len(scraped_df) == 0 or not scraped_df['Content'][0] or '403 Forbidden' in scraped_df['Content'][0] or '403 Forbidden' in scraped_df['Title'][0] :
                #Could not retrieve articles related to headline, Could possibly be a false claim OR Scraper got blocked/forbidden
                return str('Cannot retrieve articles, most likely to be false')

            scraped_content = (
            f"{scraped_df['Title'][0]} \n{scraped_df['Content'][0]}")
            answer = scraped_content
    return answer
```

```python
#To scrape headlines for the FactCC endpoint
def dataframegen(text_input):
    scraped_df = make_data(text_input)
    scraped_df.dropna(inplace=True)
    return scraped_df
```

```python
# To Run the QNA Pipeline and evaluate using FactCC
def FactCC_QNA(input_headline):
    s_scrape = time.time()
    quickSearchAnswer = google_search(input_headline)
    e_scrape = time.time()
    scrape_time = e_scrape - s_scrape

    s_factcc = time.time()
    #compare scraped info as source and headline as the claim
    ans = pipe([[[quickSearchAnswer,input_headline]]], truncation=True, padding='max_length')
    e_factcc = time.time()
    factcc_time = e_factcc - s_factcc
    if ans[0]['label'] == 'CORRECT':
        return True,quickSearchAnswer,scrape_time,factcc_time
    else:
        return False,quickSearchAnswer,scrape_time,factcc_time
```

```python
# To Run the SLM Mistral Pipeline and evaluate using FactCC
def FactCC_Mistral(input_headline ,Mistral_generated_question):
    s_scrape = time.time()
    quickSearchAnswer = google_search(Mistral_generated_question)
    e_scrape = time.time()

    #compare scraped info as source and headline as the claim
    s_factcc = time.time()
    ans = pipe([[[quickSearchAnswer,input_headline]]], truncation='only_first', padding='max_length')
    e_factcc = time.time()

    scrape_time = e_scrape - s_scrape
    factcc_time = e_factcc - s_factcc

    if ans[0]['label'] == 'CORRECT':
        return True,quickSearchAnswer,scrape_time,factcc_time
    else:
        return False,quickSearchAnswer,scrape_time,factcc_time
```

```python
# To Run the SLM Phi Pipeline and evaluate using FactCC
def FactCC_Phi(input_headline ,Phi_generated_question):

    s_scrape = time.time()
    quickSearchAnswer = google_search(Phi_generated_question)
    e_scrape = time.time()
    #compare scraped info as source and headline as the claim
    s_factcc = time.time()
    ans = pipe([[[quickSearchAnswer, input_headline]]], truncation=True, padding='max_length')
    e_factcc = time.time()

    scrape_time = e_scrape - s_scrape
    factcc_time = e_factcc - s_factcc
    if ans[0]['label'] == 'CORRECT':
        return True,quickSearchAnswer, scrape_time, factcc_time
    else:
        return False,quickSearchAnswer, scrape_time, factcc_time
```

```python
# To Run the Article Pipeline and evaluate using FactCC
def FactCC_Article(input_headline : str):
    s_scrape = time.time()
    scraped_df =  dataframegen(input_headline)
    # Convert the 'Content' column to strings
    scraped_df['Content'] = scraped_df['Content'].astype(str)

    # Sort the DataFrame based on the length of the strings in the 'Content' column
    scraped_df = scraped_df.sort_values(by='Content', key=lambda x: x.str.len(), ascending=False)

    # print(f"Title: {scraped_df['Content'][0][:100]}")
    if len(scraped_df) == 0 or not scraped_df['Content'][0] or '403 Forbidden' in scraped_df['Content'][0] or '403 Forbidden' in scraped_df['Title'][0]:
        scraped_content = "Could not scrape content"
    else:
        scraped_content = (
        f"{scraped_df['Title'][0]} \n{scraped_df['Content'][0]}")
    e_scrape = time.time()


    # Perform text classification [source,claim]
    s_factcc = time.time()
    ans = pipe([[[scraped_content,input_headline]]], truncation=True, padding='max_length')
    e_factcc = time.time()

    scrape_time = e_scrape - s_scrape
    factcc_time = e_factcc - s_factcc

    if len(scraped_df) == 0 or not scraped_df['Content'][0] or '403 Forbidden' in scraped_df['Content'][0] or '403 Forbidden' in scraped_df['Title'][0] :
    #Could not retrieve articles related to headline, Most likely be a false claim as no information is found OR Scraper got blocked/forbidden(RARE)
        return False, "could not scrape content", scrape_time, 0

    if scraped_content is None:
        scraped_content = "Could not scrape content"


    if ans[0]['label'] == 'CORRECT':
        return True, scraped_content, scrape_time, factcc_time
    else:
        return False, scraped_content, scrape_time, factcc_time
```

```python
model_zs = SummaCZS(granularity="sentence", model_name="vitc", device="cuda")  # Use GPU
model_conv = SummaCConv(models=["vitc"], bins='percentile', granularity="sentence", nli_labels="e", device="cuda", start_file="default", agg="mean")  # Use GPU
```

```python
import time
from tqdm import tqdm

# Add predictions and processing times to CSV
def add_predictions_to_csv(df, model, model_name, pipeline_name,best_threshold):
    predicted_scores = []
    true_test_labels = []
    summa_times = []  # List to store the time taken by SummaC model

    # Loop through the dataframe and process each row
    for index, row in tqdm(df.iterrows(), total=len(df)):
        document = row['Scraped Content']
        claim = row['Headline']

        # Measure the time taken for the SummaC model to process
        s_summa = time.time()
        score = model.score([str(document)], [str(claim)])
        e_summa = time.time()

        # Calculate the processing time for this row
        summa_time = e_summa - s_summa
        summa_times.append(summa_time)

        # Append the scores and labels
        true_test_labels.append(row['Actual Decision'])
        predicted_scores.append(score["scores"][0])

    # Add the predicted scores, model decisions, and processing times to the dataframe
    df['Predicted_Score'] = predicted_scores
    df[f'SummaC {model_name} Decision'] = df['Predicted_Score'].apply(lambda x: True if x >= best_threshold else False)
    df[f'SummaC {model_name} Time'] = summa_times  # Add the SummaC processing times

    # Drop the temporary 'Predicted_Score' column after use
    df = df.drop(columns=['Predicted_Score'], errors='ignore')

    # Save the updated dataframe to the CSV with | as delimiter
    df.to_csv(f'SummaC_Pipeline_{pipeline_name}.csv', index=False, sep='|')

```

```python
# Evaluation of the QNA Pipeline (FactCC)
start = time.time()

# Initialize lists to store predictions and actual values
predictions = []
actuals = []
results = []
batch_size = 2
csv_filename = 'FactCC_Pipeline_QNA.csv'
processed_headlines = set()

# Check if CSV file already exists and load processed headlines
if os.path.exists(csv_filename):
    with open(csv_filename, 'r', newline='', encoding='utf-8') as csvfile:
        csvreader = csv.reader(csvfile, delimiter='|')
        next(csvreader)  # Skip header
        for row in csvreader:
            if len(row) < 2:  # Skip rows that may not have all columns
                continue
            headline = row[1]
            processed_headlines.add(headline)

try:
    # Open the CSV file for writing with '|' as the delimiter
    with open(csv_filename, 'a', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile, delimiter='|')
        if not processed_headlines:
            # Write the header with additional columns for time tracking
            csvwriter.writerow(['Index', 'Headline', 'Scraped Content', 'Model Decision', 'Actual Decision', 'Scrape Time', 'FactCC Time'])

        # Iterate over each row in the dataframe
        for index, row in tqdm(df.iterrows(), total=len(df)):
            try:
                headline = row['Headline']
                if headline in processed_headlines:
                    print(f"Skipping already processed headline: {headline}")
                    continue

                # Get model decision, scraped content, and times
                model_decision, scraped_content, scrape_time, factcc_time = FactCC_QNA(row['Headline'])
                scraped_content_single_line = ' '.join(scraped_content.splitlines()).strip().replace('|', ' ')
                actual_decision = row['Label']
                predictions.append(model_decision)
                actuals.append(actual_decision)

                # Write the results to the CSV file
                results.append([index, row['Headline'], scraped_content_single_line, model_decision, actual_decision, scrape_time, factcc_time])

                # Write to CSV in batches of 2
                if len(results) % batch_size == 0:
                    csvwriter.writerows(results)
                    csvfile.flush()
                    os.fsync(csvfile.fileno())
                    results = []  # Reset results list after writing to CSV

                time.sleep(5)  # Delay to prevent rate limiting

            except Exception as e:
                print(f"Error processing row {index}: {str(e)}")
                continue

        # Write any remaining results
        if results:
            csvwriter.writerows(results)
            csvfile.flush()
            os.fsync(csvfile.fileno())

except Exception as e:
    print(f"An error occurred: {str(e)}")

finally:
    if os.path.exists(csv_filename):
        print(f"CSV file exists. Size: {os.path.getsize(csv_filename)} bytes")
    else:
        print("CSV file does not exist.")

end = time.time()

print(f"Total time taken: {end - start} seconds")
```

#### Barplots for time taken

```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px

# Load the data from CSV (assuming the CSV is named 'data.csv')
article = pd.read_csv('Pipeline_Article_ALL.csv', delimiter='|')
mistral = pd.read_csv('Pipeline_SLM(Mistral)_ALL.csv', delimiter='|')
phi = pd.read_csv('Pipeline_SLM(Phi)_ALL.csv', delimiter='|')
qna = pd.read_csv('Pipeline_QNA_ALL.csv', delimiter='|')

article['Pipeline'] = 'Article'
mistral['Pipeline'] = 'Mistral'
phi['Pipeline'] = 'Phi'
qna['Pipeline'] = 'QNA'

combined_df = pd.concat([article, mistral, phi, qna], ignore_index=True)

# Select the relevant time columns
time_columns = ['Scrape Time', 'FactCC Time', 'SummaC ZS Time', 'SummaC Conv Time']

y_limits = {
    'Scrape Time': (0, 18),
    'FactCC Time': (0, 0.07),
    'SummaC ZS Time': (0, 0.8),
    'SummaC Conv Time': (0, 0.7)
}

colors = {
    'Article': 'royalblue',
    'Mistral': 'darkorange',
    'Phi': 'mediumseagreen',
    'QNA': 'crimson'
}

for time_column in time_columns:
    # Create the boxplot
    fig = px.box(
        combined_df,
        x='Pipeline',
        y=time_column,
        color='Pipeline',
        color_discrete_map=colors
    )

    # Calculate the mean for each Pipeline
    means = combined_df.groupby('Pipeline')[time_column].mean().reset_index()

    # Add scatter plot for mean values on top of the boxplot
    fig.add_trace(go.Scatter(
        x=means['Pipeline'],
        y=means[time_column],
        mode='markers+text',
        marker=dict(color='black', size=7, symbol='x'),
        name='Mean',
        # text=means[time_column].round(2),  # Display rounded mean values
        # textposition='middle right'  # Position text above the mean markers
    ))

    # Update layout settings
    fig.update_layout(
        xaxis_title='Pipeline',
        yaxis_title=f'{time_column} (in seconds)',
        yaxis=dict(range=y_limits[time_column]),
        height=650,
        width=600,
        boxgap=0.35,
        boxgroupgap=0,
        showlegend=False
    )

    # Set x-axis labels
    fig.update_layout(
        xaxis=dict(tickangle=-45)
    )

    # Show the plot
    fig.show()

    print(f"Mean values for {time_column}:")
    print(means)
    print("\n")
```
